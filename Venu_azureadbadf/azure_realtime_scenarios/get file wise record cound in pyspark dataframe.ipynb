## How to track add source file name in one of column in dataframe?
* __`Answer:`__ Using `input_file_name()`  function and withColumn() function we can add new column for filenames.

```
df_airlines = spark.read.csv("/databricks-datasets/asa/airlines",header=True)
```

```
from pyspark.sql.functions import input_file_name
df_airlines = df_airlines.withColumn("FILE_NAME",input_file_name())
```

## Get no of rows on each file in a dataframe?
* __`Answer:`__ Using `input_file_name()` function and groupBy transaformation we can achive no of row on each file

```
display(df_airlines.groupBy("FILE_NAME").count())
```
